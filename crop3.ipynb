```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score

crop_database = {
    "Tomato": {
        "Description": "Tomatoes are a rich source of vitamins and antioxidants.",
        "Optimal Conditions": "Grows best in well-drained soil with plenty of sunlight.",
        "Cultivation Practices": "Plant in early spring. Water regularly and use fertilizers as needed. Protect from pests.",
        "Benefits": "High in vitamins A and C, and lycopene. Supports heart health and reduces cancer risk."
    },
    "Corn": {
        "Description": "Corn is a staple food rich in carbohydrates and fiber.",
        "Optimal Conditions": "Requires full sunlight and well-drained soil. Grows best in warmer climates.",
        "Cultivation Practices": "Plant in late spring. Ensure regular watering and use pesticides if necessary.",
        "Benefits": "Provides energy and supports digestive health. Can be used for food and animal feed."
    },
    "Wheat": {
        "Description": "Wheat is a major cereal grain that is a staple food worldwide.",
        "Optimal Conditions": "Prefers well-drained loamy soil with moderate rainfall.",
        "Cultivation Practices": "Plant in fall or spring. Requires regular watering and fertilization.",
        "Benefits": "Rich in carbohydrates and dietary fiber, supports digestive health."
    },
    "Rice": {
        "Description": "Rice is a staple food for a large part of the worldâ€™s population.",
        "Optimal Conditions": "Requires waterlogged fields and warm climate for optimal growth.",
        "Cultivation Practices": "Plant in flooded fields. Requires significant water management.",
        "Benefits": "Rich in carbohydrates and provides energy. Essential staple food."
    },
    "Potato": {
        "Description": "Potatoes are a starchy tuber that are a staple food in many diets.",
        "Optimal Conditions": "Grows best in cool climates with well-drained, loose soil.",
        "Cultivation Practices": "Plant in early spring. Regular watering and hilling are necessary.",
        "Benefits": "Rich in carbohydrates and a good source of vitamin C and potassium."
    },
    "Soybean": {
        "Description": "Soybeans are a high-protein legume used for food, oil, and animal feed.",
        "Optimal Conditions": "Grows well in well-drained soil with full sunlight.",
        "Cultivation Practices": "Plant in late spring. Requires regular watering and fertilization.",
        "Benefits": "High in protein and used in a variety of food products."
    },
    "Cotton": {
        "Description": "Cotton is a soft fiber used in textile production.",
        "Optimal Conditions": "Requires a warm climate with well-drained soil.",
        "Cultivation Practices": "Plant in late spring. Requires regular watering and pest control.",
        "Benefits": "Used in the production of textiles and clothing."
    },
    "Sugarcane": {
        "Description": "Sugarcane is a tropical grass used for producing sugar.",
        "Optimal Conditions": "Grows best in hot, sunny climates with abundant rainfall.",
        "Cultivation Practices": "Plant in early spring. Requires regular watering and fertilization.",
        "Benefits": "Primary source of sugar and used in biofuel production."
    },
    "Barley": {
        "Description": "Barley is a versatile cereal grain used for food, feed, and brewing.",
        "Optimal Conditions": "Grows well in temperate climates with well-drained soil.",
        "Cultivation Practices": "Plant in early spring or fall. Requires moderate watering.",
        "Benefits": "Rich in fiber and used in brewing and animal feed."
    },
    "Oats": {
        "Description": "Oats are a cereal grain commonly used for food and animal feed.",
        "Optimal Conditions": "Prefers cool, moist climates and well-drained soil.",
        "Cultivation Practices": "Plant in early spring. Requires regular watering.",
        "Benefits": "High in dietary fiber and used in cereals and animal feed."
    },
    "Peanut": {
        "Description": "Peanuts are a legume rich in protein and healthy fats.",
        "Optimal Conditions": "Requires sandy soil and a warm climate for optimal growth.",
        "Cultivation Practices": "Plant in late spring. Requires regular watering and hilling.",
        "Benefits": "Rich in protein, used in snacks, oil, and butter."
    },
    "Canola": {
        "Description": "Canola is a type of rapeseed used for oil production.",
        "Optimal Conditions": "Grows well in cool climates with well-drained soil.",
        "Cultivation Practices": "Plant in early spring. Requires moderate watering.",
        "Benefits": "Produces a healthy cooking oil low in saturated fat."
    },
    "Sunflower": {
        "Description": "Sunflowers are grown for their seeds and oil.",
        "Optimal Conditions": "Requires full sunlight and well-drained soil.",
        "Cultivation Practices": "Plant in late spring. Requires regular watering.",
        "Benefits": "Produces seeds rich in healthy fats and used in oil production."
    },
    "Sorghum": {
        "Description": "Sorghum is a drought-resistant cereal grain used for food, feed, and biofuel.",
        "Optimal Conditions": "Thrives in warm, dry climates with well-drained soil.",
        "Cultivation Practices": "Plant in late spring. Requires minimal watering.",
        "Benefits": "Rich in nutrients and used in cereals, animal feed, and biofuel."
    },
    "Millet": {
        "Description": "Millet is a small-seeded cereal grain used for food and animal feed.",
        "Optimal Conditions": "Grows well in arid regions with well-drained soil.",
        "Cultivation Practices": "Plant in late spring. Requires minimal watering.",
        "Benefits": "Rich in fiber and used in cereals and animal feed."
    },
    "Chickpea": {
        "Description": "Chickpeas are a legume high in protein and fiber.",
        "Optimal Conditions": "Grows best in well-drained soil with moderate rainfall.",
        "Cultivation Practices": "Plant in early spring. Requires regular watering.",
        "Benefits": "Rich in protein and fiber, used in various dishes and snacks."
    },
    "Lentil": {
        "Description": "Lentils are a high-protein legume used in soups and stews.",
        "Optimal Conditions": "Thrives in cool climates with well-drained soil.",
        "Cultivation Practices": "Plant in early spring. Requires moderate watering.",
        "Benefits": "High in protein and fiber, supports heart health."
    },
    "Flax": {
        "Description": "Flax is a plant grown for its fiber and seeds.",
        "Optimal Conditions": "Grows well in cool climates with well-drained soil.",
        "Cultivation Practices": "Plant in early spring. Requires regular watering.",
        "Benefits": "Produces seeds rich in omega-3 fatty acids and fiber."
    },
    "Alfalfa": {
        "Description": "Alfalfa is a high-protein forage crop used for animal feed.",
        "Optimal Conditions": "Grows well in well-drained soil with full sunlight.",
        "Cultivation Practices": "Plant in early spring. Requires regular watering.",
        "Benefits": "Rich in protein and used as a high-quality forage for livestock."
    },
    "Sesame": {
        "Description": "Sesame is a drought-resistant crop grown for its seeds.",
        "Optimal Conditions": "Thrives in hot, dry climates with well-drained soil.",
        "Cultivation Practices": "Plant in late spring. Requires minimal watering.",
        "Benefits": "Produces seeds rich in healthy fats and used in oil production."
    },
    "Quinoa": {
        "Description": "Quinoa is a highly nutritious grain used as a food crop.",
        "Optimal Conditions": "Grows well in cool climates with well-drained soil.",
        "Cultivation Practices": "Plant in early spring. Requires regular watering.",
        "Benefits": "High in protein, fiber, and essential amino acids."
    },
    "Cabbage": {
        "Description": "Cabbage is a leafy vegetable rich in vitamins and fiber.",
        "Optimal Conditions": "Grows best in cool climates with well-drained soil.",
        "Cultivation Practices": "Plant in early spring or fall. Requires regular watering.",
        "Benefits": "Rich in vitamins C and K, supports immune health."
    },
    "Spinach": {
        "Description": "Spinach is a leafy green vegetable rich in iron and vitamins.",
        "Optimal Conditions": "Grows best in cool, moist climates with well-drained soil.",
        "Cultivation Practices": "Plant in early spring or fall. Requires regular watering.",
        "Benefits": "High in vitamins A, C, and iron, supports eye and immune health."
    },
    "Onion": {
        "Description": "Onions are a bulb vegetable used in a wide range of dishes.",
        "Optimal Conditions": "Grows best in well-drained soil with full sunlight.",
        "Cultivation Practices": "Plant in early spring. Requires regular watering.",
        "Benefits": "Rich in vitamins C and B6, supports immune health."
    },
    "Garlic": {
        "Description": "Garlic is a bulb vegetable known for its strong flavor and health benefits.",
        "Optimal Conditions": "Grows best in well-drained soil with full sunlight.",
        "Cultivation Practices": "Plant in fall or early spring. Requires moderate watering.",
        "Benefits": "Rich in antioxidants and has antimicrobial properties."
    },
    "Carrot": {
        "Description": "Carrots are a root vegetable rich in beta-carotene.",
        "Optimal Conditions": "Grows best in cool climates with well-drained, sandy soil.",
        "Cultivation Practices": "Plant in early spring. Requires regular watering.",
        "Benefits": "Rich in vitamin A, supports eye health."
    },
    "Peas": {
        "Description": "Peas are a legume rich in protein and fiber.",
        "Optimal Conditions": "Grows best in cool climates with well-drained soil.",
        "Cultivation Practices": "Plant in early spring. Requires regular watering.",
        "Benefits": "High in protein and fiber, supports digestive health."
    },
    "Pumpkin": {
        "Description": "Pumpkins are a large fruit rich in vitamins and antioxidants.",
        "Optimal Conditions": "Requires warm climate and well-drained soil.",
        "Cultivation Practices": "Plant in late spring. Requires regular watering and space to grow.",
        "Benefits": "Rich in vitamins A and C, supports immune health."
    },
    "Cucumber": {
        "Description": "Cucumbers are a refreshing vegetable rich in water and nutrients.",
        "Optimal Conditions": "Grows best in warm climates with well-drained soil.",
        "Cultivation Practices": "Plant in late spring. Requires regular watering and support.",
        "Benefits": "Low in calories and rich in vitamins K and C."
    },
    "Lettuce": {
        "Description": "Lettuce is a leafy vegetable used in salads and sandwiches.",
        "Optimal Conditions": "Grows best in cool climates with well-drained soil.",
        "Cultivation Practices": "Plant in early spring or fall. Requires regular watering.",
        "Benefits": "Low in calories and rich in vitamins A and K."
    },
    "Broccoli": {
        "Description": "Broccoli is a cruciferous vegetable rich in vitamins and fiber.",
        "Optimal Conditions": "Grows best in cool climates with well-drained soil.",
        "Cultivation Practices": "Plant in early spring or fall. Requires regular watering.",
        "Benefits": "Rich in vitamins C and K, supports immune health."
    },
    "Cauliflower": {
        "Description": "Cauliflower is a cruciferous vegetable used in a variety of dishes.",
        "Optimal Conditions": "Grows best in cool climates with well-drained soil.",
        "Cultivation Practices": "Plant in early spring or fall. Requires regular watering.",
        "Benefits": "Rich in vitamins C and K, supports immune health."
    },
    "Beetroot": {
        "Description": "Beetroot is a root vegetable rich in nutrients and antioxidants.",
        "Optimal Conditions": "Grows best in cool climates with well-drained soil.",
        "Cultivation Practices": "Plant in early spring or fall. Requires regular watering.",
        "Benefits": "Rich in folate and manganese, supports heart health."
    },
    "Eggplant": {
        "Description": "Eggplant is a versatile vegetable used in various cuisines.",
        "Optimal Conditions": "Grows best in warm climates with well-drained soil.",
        "Cultivation Practices": "Plant in late spring. Requires regular watering and pest control.",
        "Benefits": "Rich in fiber and antioxidants, supports heart health."
    }
}


df_extended = pd.read_csv('extended_crop_data_30crops.csv')
# print(df_extended.info())
# print(df_extended.describe())

#crop analysis systim
df_extended['recommended_crop'] = df_extended['recommended_crop'].astype('category')

X_crop_extended = df_extended[['soil_type', 'climate', 'region', 'soil_pH', 'avg_temp', 'rainfall', 'market_demand']]
y_crop_extended = df_extended['recommended_crop'].cat.codes  

X_crop_train, X_crop_test, y_crop_train, y_crop_test = train_test_split(X_crop_extended, y_crop_extended, test_size=0.2, random_state=42)

crop_recommendation_model_extended = RandomForestClassifier(n_estimators=50, max_depth=10, random_state=42)
crop_recommendation_model_extended.fit(X_crop_train, y_crop_train)

# crop_accuracy_extended = accuracy_score(y_crop_test, crop_recommendation_model_extended.predict(X_crop_test))
# print(f"Extended Crop Recommendation Model Accuracy with 30 Crops: {crop_accuracy_extended}")

#yield analysis systim
X_yield_extended = df_extended[['soil_type', 'climate', 'fertilizer_used', 'irrigation', 'soil_pH', 'avg_temp', 'rainfall', 'market_demand']]
y_yield_extended = df_extended['yield']

X_yield_train, X_yield_test, y_yield_train, y_yield_test = train_test_split(X_yield_extended, y_yield_extended, test_size=0.2, random_state=42)

yield_prediction_model_extended = LinearRegression()
yield_prediction_model_extended.fit(X_yield_train, y_yield_train)

# yield_accuracy_extended = yield_prediction_model_extended.score(X_yield_test, y_yield_test)
# print(f"Extended Yield Prediction Model Accuracy: {yield_accuracy_extended}")

def recommend_crop(soil_type, climate, region, soil_pH, avg_temp, rainfall, market_demand):
    prediction = crop_recommendation_model_extended.predict([[soil_type, climate, region, soil_pH, avg_temp, rainfall, market_demand]])
    return df_extended['recommended_crop'].cat.categories[prediction[0]]

def predict_yield(soil_type, climate, fertilizer_used, irrigation, soil_pH, avg_temp, rainfall, market_demand):
    prediction = yield_prediction_model_extended.predict([[soil_type, climate, fertilizer_used, irrigation, soil_pH, avg_temp, rainfall, market_demand]])
    return prediction[0]


def get_crop_info(crop_name):
    crop_info = crop_database.get(crop_name)
    if crop_info:
        return (f"Description: {crop_info['Description']}\n"
                f"Optimal Conditions: {crop_info['Optimal Conditions']}\n"
                f"Cultivation Practices: {crop_info['Cultivation Practices']}\n"
                f"Benefits: {crop_info['Benefits']}")
    else:
        return "Crop information not found."

crop_name = input("Enter the name of the crop: ")
print(get_crop_info(crop_name))

soil_type = int(input("Enter the soil type (1, 2, 3): "))
climate = int(input("Enter the climate type (1, 2, 3): "))
region = int(input("Enter the region (1, 2): "))
fertilizer_used = int(input("Enter if fertilizer is used (1 for yes, 0 for no): "))
irrigation = int(input("Enter if irrigation is used (1 for yes, 0 for no): "))
soil_pH = int(input("Enter the pH of the soil: "))
avg_temp = int(input("Enter average temp: "))
rainfall = int(input("Enter rainfall(in mm): "))
market_demand = 2

recommended_crop = recommend_crop(soil_type, climate, region, soil_pH, avg_temp, rainfall, market_demand)
print(f"Recommended Crop based on your inputs: {recommended_crop}")

predicted_yield = predict_yield(soil_type, climate, fertilizer_used, irrigation, soil_pH, avg_temp, rainfall, market_demand)
print(f"Predicted Yield: {predicted_yield:.2f} tons/hectare")
```
